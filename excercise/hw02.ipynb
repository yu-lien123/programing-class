## 摘要

本報告旨在透過網路爬蟲技術，收集 Rotten Tomatoes 網站上資料，為近期觀影提供參考。

## 引言

電影百百種，影評網站更是多樣，這次挑選權威影評網站「爛番茄」作為觀影指南

本次爬蟲的目標是獲取近期院線片的資訊，報含海報、名稱、評價等，為觀眾提供決策支持。

## 方法

- 目標網站描述
  - 目標網站：爛番茄 首頁。  
  https://www.rottentomatoes.com/
  - 頁面結構：包含各項輪盤、電影、評價等資訊。

- 工具與技術
  - 使用 Python 的 `requests` 庫發送 HTTP 請求。
  - 使用 `BeautifulSoup` 解析 HTML。
  




```python
# 爬取資料程式碼
import requests
from bs4 import BeautifulSoup
#Step 1：Get the web content
url1 = 'https://www.rottentomatoes.com/'
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"}

response1 = requests.get(url1,headers=headers)
soup = BeautifulSoup(response1.text, "html.parser")

```

```python
#Step 2：Get the elements
from IPython.display import  Image, display, SVG
sec = soup.find('div',attrs ={'data-curation':"rt-hp-poster-list-now-playing"})

titles = sec.find_all('span', class_='p--small')
scores = sec.find_all('rt-text', attrs={"slot": "criticsScore"})
images = sec.find_all('rt-img', attrs={"slot": "image"})
scores[0].text

#Step 3：Get the content of each elements and print out
for title, score, img in zip(titles, scores, images):

  print(f"電影: {title.text.strip()}")
  score_text = score.text.strip()
  if score_text:
          print(f"新鮮度: {score_text}")
  else:
          print("評分: 暫無評分")

  display(Image(url=img.get('src')))

  print("-" * 30)
```

- 數據存儲方式
  - 寫一個csv檔儲存電影名稱與評價，以及電影海報網址

```python
# 展示資料程式碼
import csv
import requests
import os
# 爬取資料程式碼
import requests
from bs4 import BeautifulSoup
url1 = 'https://www.rottentomatoes.com/'
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"}

response1 = requests.get(url1,headers=headers)
soup = BeautifulSoup(response1.text, "html.parser")

sec = soup.find('div',attrs ={'data-curation':"rt-hp-poster-list-now-playing"})
images = sec.find_all('rt-img', attrs={"slot": "image"})
titles = sec.find_all('span', class_='p--small')
scores = sec.find_all('rt-text', attrs={"slot": "criticsScore"})


csv_filename = '/content/sample_data/movies.csv'
os.makedirs('/content/sample_data', exist_ok=True)
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
  writer = csv.writer(csvfile)

  writer.writerow(['電影名稱', '評價', '圖片網址'])
  for title, score, img in zip(titles, scores, images):
    movie_name = title.text.strip()
    score_text = score.text.strip()
    if score_text:
      movie_score = score.text.strip()
    else:
      movie_score = "暫無評分"
    img_url = img.get('src')
    writer.writerow([movie_name, movie_score, img_url])



```
