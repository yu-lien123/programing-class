```python
import random

dict = {"石頭":0, "剪刀":2, "布": 5 }

player_in = input('請輸入剪刀、石頭、布\n')
player = dict[player_in]

x = random.choice(["剪刀","石頭","布"])
com = dict[x]

print(f'你出:{player_in}')
print(f'電腦出:{x}')

if player == com:
  print("draw")
elif (player==0 and com==2) or(player==2 and com==5) or (player==5 and com==0):
  print('you win')
else:
  print('you lose')

```
