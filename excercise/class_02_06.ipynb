```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "https://www.atmovies.com.tw/movie/next/"
responses = requests.get(url)
responses.encoding = 'utf-8'
soup = BeautifulSoup(responses.text , "lxml")
soup.title
```

```python
len(soup.find_all("img" , class_="filmListAllXPoster"))
```

```python
import os

if not os.path.exists("/content/movie"):
  os.mkdir("/content/movie")

movies = [['電影','上映日期']]
for i in soup.find("ul",class_="filmListAllX").find_all("li"):
  ttl = i.img['alt']
  src = i.img['src'].strip()
  dat = i.find("div", class_ = "runtime").text
  dat = dat[5:15]
  movies.append([ttl,dat])
  print(ttl,src,dat)
  responses = requests.get(src)
  with open(f'/content/movie/{ttl}.jpg', "wb") as f:
    f.write(responses.content)
with open('/content/movies.csv','w',newline='') as f:
  w = csv.writer(f)
  w.writerows(movies)


```
