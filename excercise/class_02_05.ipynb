```python
def get_page(url):
    responses  = requests.get(url, cookies = cookies)
    soup = BeautifulSoup(responses.text, "lxml")
    for i in soup.find_all("div",class_= 'r-ent'):
      if i.find("div",class_ = "title").a:
        ttle = i.find("div",class_ = "title").a.text
        hrl = "https://www.ptt.cc" + i.find("div",class_ = "title").a.get('href')
      else :
        ttle = i.find("div",class_ = "title").text
        hrl = ""
      date = i.find("div",class_ = "date").text
      print(date,ttle,hrl)
    url = "https://www.ptt.cc" + soup.find_all("a",class_ = "btn wide")[1].get('href')
    return url

url = "https://www.ptt.cc/bbs/Gossiping/index.html"
cookies = {'over18': '1'}
for i in range(5):
  url = get_page(url)


```
